```python
%matplotlib inline
import numpy as np
import pandas as pd
"""Pandas is a python library that makes handling tabular data easier.
'Data Frames' and 'Series' ~ that allow to slice and slice rows and columns of information.
Pandas provides in-memory 2d table object called dataframe.
Pandas head() method is used to return top n (5 by default) rows of data frame or series.
"""
df = pd.read_csv("data.csv")
df.head()
```

```python

"""head() is a handy way to visualize what is loaded. We can pass it an integer
to see some specific number of rows at the beginning of DataFrame"""
df.head(10) 
```

```python
df.tail(4) # to view end of data
```

```python
df.shape # to know about dimensions
```

```python
df.size # The total size of the data frame is the rows * columns.
```

```python
len(df) # len() function gives no. of rows in a DataFrame.
```

```python
df.columns # if DataFrame has named columns(extracted automatically from the first row of a .csv file), it will give an array of them back.
```

```python
"""Extracting a single column from DataFrame, it gives bacl a "Series" in Pandas."""
df['line'] 
```

```python
"""To extract a given range of rows from a named column"""
df['line'][:20]
```

```python
# To extract a single value from a specified column/ row combination.
df['line'] [6]
```

```python
# To extract more than one column
df[['policyID', 'line']]
```

```python
# To extract specific ranges of rows from more than one column.
df[['policyID', 'line']][:10]
```

```python
# Sorting DataFrames by a specific column.
df.sort_values(['point_granularity'])
```

```python
"""To break down the number of unique values in a given column into
Series using value_counts(): distribution of data."""
counts = df['line'].value_counts()
counts
```

```python
counts.plot(kind='bar')
```
